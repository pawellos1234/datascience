```python
def parse_file(filepath):
    data = []
    with open(filepath, "r") as f:
        for line in f:
            data.append(line)
    
    return data
```

```python
def parse_file(filepath):
    with open(filepath, "r") as f:
        for line in f:
            yield line.split(';')

```

```python
to_procesed = "../datasets/otomoto.csv"
lines = parse_file(to_procesed)
header = next(lines)

def clear_price(line):
    return float(line[27].replace('"', ''))

first_line = next(lines)

count_all = 0
sum_all = 0

for line in lines:
    try:
        price = clear_price(line)
        sum_all = sum_all + price
        count_all = count_all + 1
    except Exception as e:
        continue

```

```python
import xlrd

def parse_xls(filepath):
    return xlrd.open_workbook(filename=filepath)
def get_spreadsheet(workbook, name):
    return workbook.sheet_by_name(name)

def get_cell_value(sheet, col, row):
    return sheet.cell_value(row, col)


def print_value_from_file(filename):
    print("file {}".format(filename))
    workbook = parse_xls(filename)
    sheet = get_spreadsheet(workbook, "TABLICA 1  ")
    #print("Rpws count {}".format(sheet.nrows))
    #print("columns count {}".format(sheet.ncols))
    print("Value {}".format(get_cell_value(sheet, 1, 11)))

files_list = [
    "../datasets/budzet/Sprawozdanie_operatywne_{}_{}.xlsx".format(month, year)
        for year in [2018]
        for month in (str(x).zfill(2) for x in range(1, 10))   
]
    
for filename in files_list:
    print_value_from_file(filename)
```
